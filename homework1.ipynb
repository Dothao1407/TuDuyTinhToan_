Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt

```python
a=input()
if a=="A":
    print("z")
elif a.isupper() and len(a)==1 and ord("A")<=ord(a)<=ord("Z") :
    vt=ord(a)-1
    c=chr(vt)
    print(c.lower())
else:
    print("không phải chữ cái in hoa")
```

```python
a=float(input())
b=float(input())
c=float(input())
if a+b>c and a+c>b and b+c>a:
    p=(a+b+c)/2
    s=(p*(p-a)*(p-b)*(p-c))**0.5
    print(round(s,1))
else:
    print("không phải 3 cạnh của tam giác")
```

9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
a=int(input())
if 0<=a<=9:
    match a:
        case 1:
            print("một")
        case 2:
            print("hai")
        case 3:
            print("ba")
        case 4:
            print("bốn")
        case 5:
            print("năm")
        case 6:
            print("sáu")
        case 7:
            print("bảy")
        case 8:
            print("tám")
        case 9:
            print("chín")
else:
    print("ngoài phạm vi")
```
