Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ. Nếu x là số chẵn, in ra "Even". Nếu x
là số lẻ, in ra "Odd".

```python
a=int(input("Nhập một số nguyên dương:"))
if a%2==0:
    print("Even")
else:
    print("Odd")
```

Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No"

```python
a, b=map(float,input().split())
if a<0 and b<0:
    print("Yes")
else:
    print("No")
```

Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False".

```python
a, b=list(input().split())
if len(a)>len(b):
    print("True")
else:
    print("False")
```

Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm
chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No"

```python
n=int(input())
if n%100 !=0 and n%4==0 or n%400==0:
    print("Yes")
else:
    print("No")
```

Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh

```python
a=int(input("Nhập vào 1 số điện:"))
if 0<=a<=50:
    td=a*1500
    print(td)
elif 51<=a<=100:
    td=50*1500+(a-50)*2000
    print(td)
elif a>100:
    td=50*1500+50*2000+(a-100)*3000
    print(td)


      
```

Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc:
- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu

```python
a=float(input("Nhập điểm trung bình của 1 học sinh"))
if a>=8:
    print("giỏi")
elif 6.5<=a<8:
    print("Khá")
elif 5<=a<6.5:
    print("Trung bình")
elif a<5:
    print("Yếu")
```
