W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
(từ 2 đến căn của n nếu n chia hết cho 1 số trong khoảng thì không phải số nguyên tố)

```python
import math 
def kiemtrasonguyento(n:int)->bool:
    if n<2:
        return False 
    for i in range(2,int(math.sqrt(n)+1)):
        if n%i==0:
            return False
    return True
n=int(input())
print(kiemtrasonguyento(n))
```

```python
import math
n=int(input())
def kiemtrasonguyento(n):
    if n<2:
        return False
    for i in range (2,int(math.sqrt(n)+1)):
        if n%i==0:
            return False
    return True
print(kiemtrasonguyento(n))
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def kiemtrasohoanhao(n:int)->bool:
    s=0
    if n<=1:
        return False
    for i in range (1,n):
        if n%i==0:
            s+=i
    if s==n:
        return True
    else:
        return False
n=int(input())
print(kiemtrasohoanhao(n))
    
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
a, b, op=input().split()
a=float(a)
b=float(b)
def compute(a,b,op):
    if op=='+':
        return a+b
    if op=='-':
        return a-b
    if op=='*':
        return a*b
    if op=='/':
        return a/b
print(compute(a,b,op))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
a, b, c=map(float,input().split())
def is_right_triangle(a,b,c):
    if a>0 and b>0 and c>0 and a+b>c and b+c>a and a+c>b and a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        return True
    else:
        return False
print(is_right_triangle(a,b,c))
```
