W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
def la_so_mu_cua_2(n):
    if n <= 0:
        return False
    while n % 2 == 0:
        n //= 2
    return n == 1
n=int(input())
print(la_so_mu_cua_2(n))

```

W4A23
Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập n: "))
tong = 0
k = 0

while tong + (k+1) < n:
    k += 1
    tong += k

print("k =", k, "với S(k) =", tong)
```

W4A24
Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A


```python
A = float(input("Nhập A: "))
tong = 0
n = 0

while tong <= A:
    n += 1
    tong += 1/n

print("n nhỏ nhất là:", n)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập


```python
ds = []

while True:
    x = int(input("Nhập số (nhập -1 để dừng): "))
    if x == -1:
        break
    ds.append(x)

if ds: 
    print("Số lớn nhất:", max(ds))
    print("Số nhỏ nhất:", min(ds))
else:
    print("Không có số nào được nhập.")
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1

while f2 <= A:
    f1, f2 = f2, f1 + f2

print("Số Fibonacci lớn nhất không vượt quá A là:", f1)
```

W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
chuoi = input("Nhập chuỗi: ")
dem = 0
trong_tu = False

for i in range(len(chuoi)):
    if chuoi[i] != " " and not trong_tu:
        dem += 1
        trong_tu = True
    elif chuoi[i] == " ":
        trong_tu = False

print("Số từ trong chuỗi là:", dem)
```

W4A28
Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split() 


```python
chuoi = input("Nhập chuỗi: ")
tu_dau = ""

for i in range(len(chuoi)):
    if chuoi[i] == " ":
        break
    tu_dau += chuoi[i]

print("Từ đầu tiên là:", tu_dau)
```

W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa

```python
chuoi = input("Nhập chuỗi: ")
ket_qua = ""

for ch in chuoi:
    if ch.islower():        # nếu là chữ thường
        ket_qua += ch.upper()
    elif ch.isupper():      # nếu là chữ hoa
        ket_qua += ch.lower()
    else:
        ket_qua += ch       # phòng trường hợp ký tự khác

print("Chuỗi sau khi đổi:", ket_qua)
```

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
chuoi = input("Nhập chuỗi: ")

dem_hoa = 0
dem_thuong = 0
dem_so = 0

for ch in chuoi:
    if ch.isupper():
        dem_hoa += 1
    elif ch.islower():
        dem_thuong += 1
    elif ch.isdigit():
        dem_so += 1

print("Số ký tự in hoa:", dem_hoa)
print("Số ký tự in thường:", dem_thuong)
print("Số ký tự số:", dem_so)
```
